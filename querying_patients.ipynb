```python
# import statements for the software packages and modules we will use in this tutorial
import os
import numpy as np
import pandas as pd
import plotnine
from plotnine import *
import pandas_gbq
from datetime import datetime
from google.cloud import bigquery

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# CKD 3, 4, 5, ESRD Patients and their disease history
query = f"""
WITH target_cohort AS (
    SELECT DISTINCT person_id
    FROM {dataset}.condition_occurrence
    WHERE condition_concept_id IN (443597, 443612, 443611, 193782)
)

SELECT 
    co.person_id,
    -- 1. Disease Name
    c1.concept_name AS condition_name,
    co.condition_start_date,
    co.condition_end_date,
    -- 2. Type Name (e.g., "EHR Chief Complaint" or "Claim")
    c2.concept_name AS condition_type_name,
FROM {dataset}.condition_occurrence co
JOIN target_cohort tc ON co.person_id = tc.person_id
-- Join for the main disease name
JOIN {dataset}.concept c1 ON co.condition_concept_id = c1.concept_id
-- Join for the record type
LEFT JOIN {dataset}.concept c2 ON co.condition_type_concept_id = c2.concept_id
ORDER BY co.person_id, co.condition_start_date;
"""

ckd_patients_disease_history = pandas_gbq.read_gbq(query)
```

```python
# CKD 3, 4, 5, ESRD Patients and their disease history
query = f"""
WITH target_cohort AS (
    SELECT DISTINCT person_id
    FROM {dataset}.condition_occurrence
    WHERE condition_concept_id IN (443597, 443612, 443611, 193782)
)

SELECT 
    co.person_id,
    -- 1. Disease Name
    c1.concept_name AS condition_name,
    co.condition_start_date,
    co.condition_end_date,
    -- 2. Type Name (e.g., "EHR Chief Complaint" or "Claim")
    c2.concept_name AS condition_type_name,
    -- 3. Status Name (e.g., "Final diagnosis")
    c3.concept_name AS condition_status_name,
    --c4.concept_name AS source_concept_name,
    v.vocabulary_name AS source_vocabulary,
    co.condition_source_value,
    --co.condition_occurrence_id,
    co.condition_concept_id,
    co.condition_type_concept_id,
    co.condition_status_concept_id,
    c5.concept_name as visit_concept_name,
    vo.visit_concept_id,
    c6.concept_name as visit_type_concept_name,
    vo.visit_type_concept_id
FROM {dataset}.condition_occurrence co
JOIN target_cohort tc ON co.person_id = tc.person_id
-- Join for the main disease name
JOIN {dataset}.concept c1 ON co.condition_concept_id = c1.concept_id
-- Join for the record type
LEFT JOIN {dataset}.concept c2 ON co.condition_type_concept_id = c2.concept_id
-- Join for the clinical status
LEFT JOIN {dataset}.concept c3 ON co.condition_status_concept_id = c3.concept_id
-- Join 4: Source Concept Name
LEFT JOIN {dataset}.concept c4 ON co.condition_source_concept_id = c4.concept_id
-- Join 5: Vocabulary Name (e.g., identifying if it's ICD-9, ICD-10, etc.)
LEFT JOIN {dataset}.vocabulary v ON c4.vocabulary_id = v.vocabulary_id
-- Join 6: visit name and visit type
LEFT JOIN {dataset}.visit_occurrence vo ON co.visit_occurrence_id = vo.visit_occurrence_id
JOIN {dataset}.concept c5 on vo.visit_concept_id = c5.concept_id
JOIN {dataset}.concept c6 on vo.visit_type_concept_id = c6.concept_id
ORDER BY co.person_id, co.condition_start_date;
"""

ckd_patients_disease_history = pandas_gbq.read_gbq(query)
```

```python
# Control cohort and their disease history
query = f"""
WITH target_cohort AS (
    SELECT DISTINCT person_id
    FROM {dataset}.condition_occurrence
    WHERE condition_concept_id IN (443597, 443612, 443611, 193782)
),

control_pool AS (
    SELECT person_id
    FROM {dataset}.condition_occurrence
    WHERE person_id NOT IN (SELECT person_id FROM target_cohort)
    ORDER BY FARM_FINGERPRINT(CONCAT(CAST(person_id AS STRING), 'seed123'))
    LIMIT 100
)

SELECT 
    co.person_id,
    -- 1. Disease Name
    c1.concept_name AS condition_name,
    co.condition_start_date,
    co.condition_end_date,
    -- 2. Type Name (e.g., "EHR Chief Complaint" or "Claim")
    c2.concept_name AS condition_type_name,
    -- 3. Status Name (e.g., "Final diagnosis")
    c3.concept_name AS condition_status_name,
    --c4.concept_name AS source_concept_name,
    v.vocabulary_name AS source_vocabulary,
    co.condition_source_value,
    --co.condition_occurrence_id,
    co.condition_concept_id,
    co.condition_type_concept_id,
    co.condition_status_concept_id
FROM {dataset}.condition_occurrence co
JOIN control_pool tc ON co.person_id = tc.person_id
-- Join for the main disease name
JOIN {dataset}.concept c1 ON co.condition_concept_id = c1.concept_id
-- Join for the record type
LEFT JOIN {dataset}.concept c2 ON co.condition_type_concept_id = c2.concept_id
-- Join for the clinical status
LEFT JOIN {dataset}.concept c3 ON co.condition_status_concept_id = c3.concept_id
-- Join 4: Source Concept Name
LEFT JOIN {dataset}.concept c4 ON co.condition_source_concept_id = c4.concept_id
-- Join 5: Vocabulary Name (e.g., identifying if it's ICD-9, ICD-10, etc.)
LEFT JOIN {dataset}.vocabulary v ON c4.vocabulary_id = v.vocabulary_id
ORDER BY co.person_id, co.condition_start_date;
"""

non_ckd_patients_disease_history = pandas_gbq.read_gbq(query)
```
