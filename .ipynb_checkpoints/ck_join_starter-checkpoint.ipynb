# CKD/ESRD cohort starter notebook

This notebook:
1. Reads all `.csv.gz` files from `~/workspace/ckd-esrd-cohort-20250505`
2. Merges split files that share the same base table name
3. Keeps merged tables in memory as pandas DataFrames
4. Joins selected tables into one dataframe keyed by `person_id`

```python
from pathlib import Path
import re
import pandas as pd
import matplotlib.pyplot as plt

source_dir = Path('~/workspace/ckd-esrd-cohort-20250505').expanduser()
gz_files = sorted(source_dir.glob('*.csv.gz'))
print(f'Found {len(gz_files)} .csv.gz files in {source_dir}')
```

```python
def normalize_table_name(raw_name: str) -> str:
    name = re.sub(r'^\d+[_\-.]+', '', raw_name)
    name = re.sub(r'[_\-.]?\d+$', '', name)
    return name or raw_name

table_to_parts = {}
for gz_path in gz_files:
    raw_name = gz_path.name.replace('.csv.gz', '')
    table_name = normalize_table_name(raw_name)
    table_to_parts.setdefault(table_name, []).append(gz_path)

tables = {}
for table_name, parts in sorted(table_to_parts.items()):
    part_dfs = [pd.read_csv(part, low_memory=False) for part in sorted(parts)]
    combined_df = pd.concat(part_dfs, ignore_index=True, sort=False)
    tables[table_name] = combined_df

    if len(parts) > 1:
        print(f'Combined {len(parts)} parts in-memory -> {table_name} ({combined_df.shape})')
    else:
        print(f'Loaded in-memory -> {table_name} ({combined_df.shape})')

print(f'Prepared {len(tables)} merged tables in memory')
```

```python
wanted_tables = [
    'person',
    'stepsIntraday',
    'waistCircumference',
    'bloodPressure',
    'weight',
    'heigh',
    'hipCircumference',
    'heartRateSummary',
    'sleepDailySummary',
    'conditionOccurence',
    'bmi',
]

aliases = {
    'heigh': ['height'],
    'conditionoccurence': ['conditionoccurrence'],
}

table_lookup = {k.lower(): v for k, v in tables.items()}

def find_table_df(table_name: str):
    key = table_name.lower()

    if key in table_lookup:
        return table_lookup[key]

    for alt in aliases.get(key, []):
        if alt in table_lookup:
            return table_lookup[alt]

    for name, df in table_lookup.items():
        if key in name:
            return df

    for alt in aliases.get(key, []):
        for name, df in table_lookup.items():
            if alt in name:
                return df

    return None

selected_tables = {}
missing = []

for t in wanted_tables:
    df = find_table_df(t)
    if df is None:
        missing.append(t)
        continue
    selected_tables[t] = df

print('Loaded tables:', sorted(selected_tables.keys()))
if missing:
    print('Missing tables:', missing)
```

# Investigate conditionOccurrence

```python
condition_df = selected_tables['conditionOccurence']
```

## Plot first occurrence time of condition

```python
# Step 1: Select the minimum condition_start_datetime for each person_id
min_condition_df = condition_df.groupby('person_id')['condition_start_datetime'].min().reset_index()
# Step 2: Bin the condition_start_datetime into 1 month intervals
min_condition_df['month_interval'] = pd.to_datetime(min_condition_df['condition_start_datetime'], format='mixed').dt.to_period('Y')
# Step 3: Count the number of people in each month interval
monthly_counts = min_condition_df['month_interval'].value_counts().sort_index()
# Step 4: Plot the number of people in each month interval
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar')
plt.xlabel('Year Interval')
plt.ylabel('Number of People')
plt.title('Number of People by Month Interval of First Condition Start')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Plot first occurrence age of condition

```python
person_df = selected_tables['person']
person_df[['person_id','date_of_birth']].head()
```

```python
# Join person_df['date_of_birth', 'person_id'] to min_condition_df
joined_df = pd.merge(min_condition_df, person_df[['person_id','date_of_birth']], on='person_id')
# Get age at first occurrence
joined_df['age_first_diagnosed'] = pd.to_datetime(joined_df['condition_start_datetime'],format="mixed") - pd.to_datetime(joined_df['date_of_birth'], format='mixed')
joined_df['age_first_diagnosed'] = joined_df['age_first_diagnosed'] / pd.Timedelta(days=365.25)
```

```python
# Plot a histogram
plt.figure(figsize=(10, 6))
plt.hist(joined_df['age_first_diagnosed'] , bins=30, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Age First Diagnosed with CKD')
plt.show()
```

```python
conditionOccurrence_sql = r"(

    SELECT
        condition_concept_id,
        condition_end_datetime,
        condition_source_concept_id,
        condition_source_value,
        condition_start_datetime,
        condition_status_concept_id,
        condition_status_concept_name,
        T_DISP_condition_status_concept_name,
        condition_status_source_value,
        condition_type_concept_id,
        condition_type_concept_name,
        T_DISP_condition_type_concept_name,
        person_id,
        source_concept_code,
        T_DISP_source_concept_code,
        source_concept_name,
        T_DISP_source_concept_name,
        source_vocabulary,
        T_DISP_source_vocabulary,
        standard_concept_code,
        T_DISP_standard_concept_code,
        standard_concept_name,
        T_DISP_standard_concept_name,
        standard_vocabulary,
        T_DISP_standard_vocabulary,
        stop_reason,
        visit_occurrence_id 
    FROM
        `wb-silky-artichoke-2408.C2024Q3R8_index_111825`.T_ESA_conditionOccurrence_person_id 
    WHERE
        (
            (
                person_id IN (SELECT
                    person_id AS primary_id 
                FROM
                    `wb-silky-artichoke-2408.C2024Q3R8_index_111825`.T_ENT_conditionOccurrence 
                WHERE
                    (condition_concept_id IN (SELECT
                        descendant 
                    FROM
                        `wb-silky-artichoke-2408.C2024Q3R8_index_111825`.T_HAD_conditionConcept_default 
                    WHERE
                        ancestor IN (198185, 443597, 443611, 443612, 4128067, 35623051) 
                    UNION
                    ALL SELECT
                        198185 
                    UNION
                    ALL SELECT
                        443597 
                    UNION
                    ALL SELECT
                        443611 
                    UNION
                    ALL SELECT
                        443612 
                    UNION
                    ALL SELECT
                        4128067 
                    UNION
                    ALL SELECT
                        35623051)))
                )
            ) 
            AND (
                condition_concept_id IN (SELECT
                    descendant 
                FROM
                    `wb-silky-artichoke-2408.C2024Q3R8_index_111825`.T_HAD_conditionConcept_default 
                WHERE
                    ancestor IN (198185, 443597, 443611, 443612, 4128067) 
                UNION
                ALL SELECT
                    198185 
                UNION
                ALL SELECT
                    443597 
                UNION
                ALL SELECT
                    443611 
                UNION
                ALL SELECT
                    443612 
                UNION
                ALL SELECT
                    4128067)
            ))"

# Formulate a Cloud Storage destination path for the data exported from BigQuery.
# NOTE: By default data exported multiple times on the same day will overwrite older copies.
#       But data exported on a different days will write to a new location so that historical
#       copies can be kept as the dataset definition is changed.
conditionOccurrence_path <- file.path(
  "gs://ckd-esrd-cohort-20250505-wb-glacial-acorn-1350",
  "bq_exports",
  "conditionOccurrence",
  strftime(lubridate::now(), "%Y%m%d"),  # Comment out this line if you want the export to always overwrite.
  "conditionOccurrence",
  "conditionOccurrence_*.csv")
message(str_glue('The data will be written to {conditionOccurrence_path}. Use this path when reading ',
                 'the data into your notebooks in the future.'))

# Perform the query and export the dataset to Cloud Storage as CSV files.
# NOTE: You only need to run `bq_table_save` once. After that, you can
#       just read data from the CSVs in Cloud Storage.
bq_table_save(
  bq_dataset_query("wb-silky-artichoke-2408.C2024Q3R8_index_111825", conditionOccurrence_sql, billing = "wb-glacial-acorn-1350"),
  conditionOccurrence_path,
  destination_format = "CSV")

# Read the data directly from Cloud Storage into memory.
# NOTE: Alternatively you can `gsutil -m cp {conditionOccurrence_path}` to copy these files
#       to the Jupyter disk.
read_bq_export_from_workspace_bucket <- function(export_path) {
  col_types <- NULL
  bind_rows(
    map(system2('gsutil', args = c('ls', export_path), stdout = TRUE, stderr = TRUE),
        function(csv) {
          message(str_glue('Loading {csv}.'))
          chunk <- read_csv(pipe(str_glue('gsutil cat {csv}')), col_types = col_types, show_col_types = FALSE)
          if (is.null(col_types)) {
            col_types <- spec(chunk)
          }
          chunk
        }))
}

conditionOccurrence_df <- read_bq_export_from_workspace_bucket(conditionOccurrence_path)
dim(conditionOccurrence_df)
head(conditionOccurrence_df, 5)
```
